```python
img = cv.imread(r"Resources\Exercise\Lenna_(test_image).png")
cv.imshow('Lenna', img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.imshow(img)
```

```python
part1 = img[:img.shape[0]//2, :img.shape[1]//2]
plt.imshow(part1)
```

```python
part2 = img[:img.shape[0]//2, img.shape[1]//2:]
plt.imshow(part2)
```

```python
part3 = img[img.shape[0]//2:, :img.shape[1]//2]
plt.imshow(part3)
```

```python
part4 = img[img.shape[0]//2:, img.shape[1]//2:]
plt.imshow(part4)
```

```python
row_1 = cv.hconcat([part3,part1])
plt.imshow(row_1);
```

```python
row_2 = cv.hconcat([part4,part2])
plt.imshow(row_2);
```

```python
img_mix = cv.vconcat([row_1, row_2])
plt.imshow(img_mix);
```
